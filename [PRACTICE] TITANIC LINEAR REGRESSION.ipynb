```python
data_drop=data.drop(['body', 'home.dest','cabin','boat','Unnamed: 0','ticket'], axis=1)
data_drop.head()
```

```python
t=[]
for i in data_drop['pclass']:
    if 'st' in i:
        m=i.split('s')
    if 'rd' in i:
        m=i.split('r')
    if 'nd' in i:
        m=i.split('n')
    t.append(m[0])
data_drop['PClass']=t
```

```python
data_drop=data_drop.drop('pclass',axis=1)
data_drop=data_drop.drop('name',axis=1)
data_drop.head()
```

```python
data_drop['age']=data_drop['age'].fillna(data_drop['age'].mean())
data_drop
```

```python
data_drop=data_drop.dropna(axis=0)
data_drop.info()
```

```python
Embarked=[]
for i in data_drop['embarked']:
    if i=='Southampton':
        Embarked.append(1)
    if i=='Cherbourg':
        Embarked.append(2)
    if i=='Queenstown':
        Embarked.append(3)
data_drop['embarked']=Embarked
```

```python
Sex=[]
for i in data_drop['sex']:
    if i=='female':
        Sex.append(0)
    if i=='male':
        Sex.append(1)
data_drop['sex']=Sex
```

```python
data_drop['age'] = data_drop['age'].astype(int)
data_drop['PClass'] = data_drop['PClass'].astype(int)
data_drop['fare'] = data_drop['fare'].astype(int)
y=np.array([data_drop['survived']]).T
x=np.array([data_drop['sex'],data_drop['age'],data_drop['sibsp'],data_drop['parch'],data_drop['fare'],data_drop['embarked'],data_drop['PClass']]).T
```

```python
one=np.ones((x.shape[0],1))
Xbar= np.concatenate((one, x), axis = 1)
```

```python
A=np.dot(Xbar.T,Xbar)
b=np.dot(Xbar.T,y)
w=np.dot(np.linalg.pinv(A),b)
w
```

```python
from sklearn.metrics import mean_squared_error
y_true=y
y_pred=np.dot(Xbar,w)
mean_squared_error(y_true,y_pred)
```
