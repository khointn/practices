```python
def word(meaning,value_eng):
    count=0
    for i in meaning:
        if i in value_eng.lower():
            count+=1
    if count/len(value_eng)>=0.9 and count/len(value_eng)<=1.1:
        return True
    else:
        return False
```

```python
wordsperday=100
```

```python
count_dung=0
listkey_eng=random.sample(list(dict_sat.keys()),wordsperday)
key_sai=[]
for key_eng in listkey_eng:
    value_eng=dict_sat.get(key_eng)
    print(key_eng)
    meaning=input()
    if word(meaning,value_eng)==True:
        print('bạn đúng rồi')
        count_dung+=1
    else:
        print('sai rồi huhu. Nghĩa đúng là: ',value_eng)
        key_sai.append(key_eng)
print ('Tốt quá hurrayyy !!!' if count_dung/wordsperday >=0.8 else 'Hong ổn rồi, học lại nào', '. Tỉ lệ đúng là: ',count_dung/wordsperday)
print('HỌC LẠI !!!!')
######## Re-do the wrong answer repeatedly until correct all#######
while count_dung/wordsperday !=1:
    for key_eng in key_sai:
        value_eng=dict_sat.get(key_eng)
        print(key_eng)
        meaning=input()
        if word(meaning,value_eng)==True:
            print('bạn đúng rồi')
            count_dung+=1
            key_sai.remove(key_eng)
        else:
            print('sai rồi huhu. Nghĩa đúng là: ',value_eng)
print('Awww finally finished!!! Well done Khôi')
```
