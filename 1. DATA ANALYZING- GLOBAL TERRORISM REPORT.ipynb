```python
#plotted a pie chart( little bit ugly)
colors=['gray','orange','b']
label=['Afghanistan','Pakistan','Turkmenistan']
plt.pie(data[data.gname=='Taliban'].country_txt.value_counts(), labels=label,colors=colors, startangle=180, autopct='%.1f%%')
plt.title('Active range (countries) of Taliban')
plt.show()
```

```python
#By Attacktypes:
data[data.gname=='Taliban']['attacktype1_txt'].value_counts()
#Plotted pie chart:
fig, (ax1, ax2) = plt.subplots(1, 2)
dothi=[2866,2125,713,508,235,33+19+13+966]
label=['Bombing/Explosion','Armed Assault','Kidnapping','Assassination','Facility/Infrastructure Attack','Others+Unknown']
ax1.pie(dothi, startangle=180, autopct='%.1f%%')
#Compared attack types of Taliban to attack types of global:
data['attacktype1_txt'].value_counts()
#Plotted pie chart:
dothi1=[88255,42669,19312,11158,10356,7276+1015+991+659]
ax2.pie(dothi1, startangle=180, autopct='%.1f%%')
plt.legend(label,loc='lower right',bbox_to_anchor=(1, 1),fontsize=8)
plt.show()

#Based on the charts below:
#attacktype statisticof Taliban is nearly the same to the world's. 
#This was reasonable because Taliban were org with most attacks done
#Taliban was least possible to use Bombing in an attack compare to the world in general, instead it was Armed Assault
```

```python
#By size:
#For simple, we calculate the damage of an attack by the method (I self created): damage=death+ wounded
data_phantich['criticalrate']=data_phantich['death']+data_phantich['wound']
#drop the nulls (some did not have death and some did not have wound, but it was a very small number compared to the analyzing one:
datadeath=data_phantich.dropna(axis=0,subset=['death','wound','criticalrate'])
#Calculated the average in damage of Taliban compared to the world's
datadeath[datadeath.terrorname=='Taliban']['criticalrate'].mean()#==7.9109529343405
datadeath['criticalrate'].mean() #==5.296128433353355
#=> Taliban cÃ³ má»©c Ä‘á»™ thiá»‡t háº¡i cao hÆ¡n so vá»›i má»©c trung bÃ¬nh toÃ n cáº§u trong 1 vá»¥ táº¥n cÃ´ng 
#=> In average, Taliban's attack caused higher damage compared to world's
data_phantich['deathrate']=data_phantich['death']/(data_phantich['death']+data_phantich['wound'])
datadeath['deathrate']=datadeath['death']/(datadeath['death']+datadeath['wound'])
datadeath['deathrate'].fillna(value=0,axis=0,inplace=True)
#Calculate the death rate of Taliban's attack compared to the world's:
datadeath[datadeath.terrorname=='Taliban']['deathrate'].mean() #0.512434661134669
datadeath['deathrate'].mean() #0.32942326312248466
```

```python
#Sum up the attack size of Taliban:
#AFFECTED NUMBER (DEATH + WOUNDED):
    #On average, in an attack caused by Taliban, the affected people were 8
    #Higher than the world's average (5 people)
#DEATH RATE (DEATH/AFFECTED NUMBER)
    # If you were in an Taliban's attack, you would have 51,24% of death chance !!!
    #In other way, you just have 48,76% survival chance!!!
    #Extremely higher than a random attack from world- only 32,94% death rate and 67,06% survive!!!
    
#==> Taliban was extremely dangerous!
```
